```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('data/final/final_df.csv')
df2 = pd.read_csv('data/lstm_predictions.csv')
```

```python
df2.rename(columns={
    "pred_next_games_played_per": 'Predicted Games Played Percentage LSTM',
    "pred_next_goals_per_game": 'Predicted Goals Scored Percentage LSTM',
    "pred_next_assists_per_game": 'Predicted Assists Percentage LSTM'
}, inplace=True)
```

```python
df = pd.merge(df1, df2, on="playerId", how="inner") 
df.drop(columns=["Unnamed: 0.1", "Unnamed: 0"], inplace=True)
df.head()
```

```python
df['Predicted Games Played LSTM'] = np.ceil(df['Predicted Games Played Percentage LSTM'] * 82)
df['Predicted Goals Scored per 82 Games LSTM'] = np.ceil(df['Predicted Goals Scored Percentage LSTM'] * 82)
df['Predicted Assits per 82 Games LSTM'] = np.ceil(df['Predicted Assists Percentage LSTM'] * 82)
```

```python
df.head()
```

```python
df.to_csv('data/final/final_df2.csv')
```
